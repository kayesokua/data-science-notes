```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# load the data into a pandas dataframe
df = pd.read_csv('data/dance/contemporary_dance_solo_20230219225812_10.csv')
landmarks = df.columns.to_list()

df_head = df[['nose_x', 'nose_y', 'nose_z', 'left_eye_x', 'left_eye_y', 'left_eye_z',
              'right_eye_x', 'right_eye_y', 'right_eye_z', 'left_ear_x', 'left_ear_y', 'left_ear_z',
              'right_ear_x', 'right_ear_y', 'right_ear_z', 'mouth_left_x', 'mouth_left_y', 'mouth_left_z',
              'mouth_right_x', 'mouth_right_y', 'mouth_right_z']]

df_torso = df[['left_shoulder_x', 'left_shoulder_y', 'left_shoulder_z', 'right_shoulder_x', 'right_shoulder_y', 'right_shoulder_z',
               'left_elbow_x', 'left_elbow_y', 'left_elbow_z', 'right_elbow_x', 'right_elbow_y', 'right_elbow_z',
               'left_wrist_x', 'left_wrist_y', 'left_wrist_z', 'right_wrist_x', 'right_wrist_y', 'right_wrist_z']]

df_hip = df[['left_hip_x', 'left_hip_y', 'left_hip_z', 'right_hip_x', 'right_hip_y', 'right_hip_z',
               'left_knee_x', 'left_knee_y', 'left_knee_z', 'right_knee_x', 'right_knee_y', 'right_knee_z',
               'left_ankle_x', 'left_ankle_y', 'left_ankle_z', 'right_ankle_x', 'right_ankle_y', 'right_ankle_z',
               'left_heel_x', 'left_heel_y', 'left_heel_z', 'right_heel_x', 'right_heel_y', 'right_heel_z',
               'left_foot_index_x', 'left_foot_index_y', 'left_foot_index_z', 'right_foot_index_x', 'right_foot_index_y', 'right_foot_index_z']]

sns.set(style="darkgrid")
plt.figure(figsize=(15,3))
cmap = sns.color_palette("coolwarm", as_cmap=True)
sns.lineplot(x=df_head.index, y='nose_x', data=df_head, label='Nose')
sns.lineplot(x=df_head.index, y='left_eye_x', data=df_head, label='Left Eye')
sns.lineplot(x=df_head.index, y='right_eye_x', data=df_head, label='Right Eye')
sns.lineplot(x=df_head.index, y='left_ear_x', data=df_head, label='Left Ear')
sns.lineplot(x=df_head.index, y='right_ear_x', data=df_head, label='Right Ear')
sns.lineplot(x=df_head.index, y='mouth_left_x', data=df_head, label='Mouth Left')
sns.lineplot(x=df_head.index, y='mouth_right_x', data=df_head, label='Mouth Right')
plt.title('Head Movements', fontsize=10)
plt.legend().remove()
plt.show()
```

## Rate of Change in Movement

```python
df_head.pct_change().plot(figsize=(15,3), title='Head Movements', cmap=cmap, legend=False)
df_torso.pct_change().plot(figsize=(15,3), title='Torso Movements', cmap=cmap, legend=False)
df_hip.pct_change().plot(figsize=(15,3), title='Hip Movements', cmap=cmap, legend=False)
```
